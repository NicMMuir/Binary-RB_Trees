```python
import matplotlib.pyplot as plt
```

# Insertion of nodes

```python
arrheight = []
arrcount = []
arrtime = []

with open('BinaryTreeInsert.txt') as file:
    count = 0
    temp = [0.0,0.0,0.0]
    for line in file:
        count += 1
        ln = line.split(" ");
        temp[0] += float(ln[0])
        temp[1] += float(ln[1])
        temp[2] += float(ln[2])
        
        if count == 3:
            arrcount.append(temp[0]/3)
            arrtime.append(temp[1]/3)
            arrheight.append(temp[2]/3)
            temp = [0.0,0.0,0.0]
            count = 0
            
            

print(arrcount)
print(arrtime)
print(arrheight)
        
```

```python
plt.title("Binary Tree Insert Time vs Height")
plt.plot(arrcount,arrheight)
plt.xlabel("Height")
plt.ylabel("Number of nodes")
plt.savefig("BT_TimevsHeight.jpg")
plt.show()
```

```python
plt.title("Binary Tree Tree Time vs Insertion of Nodes")
plt.plot(arrcount,arrtime)
plt.xlabel("Number of Nodes")
plt.ylabel("Time (Nanoseconds)")
plt.savefig("BT_TimevsNN.jpg")
plt.show()
```

# Deletion of nodes

```python
arrcount = []
arrtime = []


with open('BinaryTreeDelete.txt') as file:
    count = 0
    temp = [0.0,0.0]
    for line in file:
        count += 1
        ln = line.split(" ");
        temp[0] += float(ln[0])
        temp[1] += float(ln[1])

        
        if count == 3:
            arrcount.append(temp[0]/3)
            arrtime.append(temp[1]/3)
            
            temp = [0.0,0.0]
            count = 0
            
            

print(arrcount)
print(arrtime)
```

```python
plt.title("Binary Tree Tree Time vs Deletion of Nodes")
plt.plot(arrcount,arrtime)
plt.xlabel("Number of Nodes")
plt.ylabel("Time (Nanoseconds)")
plt.savefig("BT_Delete_TimevsNN.jpg")
plt.show()
```
